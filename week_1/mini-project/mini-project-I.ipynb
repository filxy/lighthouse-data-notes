## Task
How many tube and bus lines are in London? Print names of all tube lines.

````python
url = "https://api.tfl.gov.uk/Line/Route"
response = re.get(url)
JSON(response.json())
json_resp = response.json()
JSON(json_resp)
````

````python
BusCount = 0
TubeCount = 0
List=[]
for line in json_resp:
    if(line["modeName"]=="bus"):
        BusCount+=1
    elif(line["modeName"]=="tube"):
        TubeCount+=1
        List.append(line["name"])
        
print(f'number of bus lines in London are : {BusCount}')
print(f'number of tube lines in London are : {TubeCount}')
print(f'names of tube lines:{List}')
````
